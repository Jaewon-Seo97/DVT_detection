```python
###########select min,max
al_max=[]
al_min=[]
center_top= []
center_bottom= []
for imfl in range(len(imgfol_list)):
# for imfl in range(0,1):
    imgfile_list=sorted(glob.glob(imgfol_list[imfl]+'/*.png'))
#     print(imgfile_list, len(imgfile_list))
#     os.mkdir(crop_path+'/'+imgfol_name[imfl])
    if len(imgfile_list)==0:
        print(imgfol_list[imfl])

    for ilist in range(len(imgfile_list)): 
#     for ilist in range(0,1): 

        top_loc=[]
        img= imread(imgfile_list[ilist], as_gray=True)
        top=img.copy()

        top[top > 0] = 255
        top[top <= 0] = 0
        
        morp= morphology.binary_opening(top, morphology.diamond(3)).astype(np.uint8)
        
        hole=ndimage.binary_fill_holes(morp).astype(int)
        
        hole[hole > 0] = 255
        hole[hole <= 0] = 0
        
#         plt.figure(figsize=(10,10))
#         plt.imshow(hole, cmap='gray')
#         print(hole)
        xx = np.where(hole == 255)
        al_max.append(np.max(xx[0]))
        al_min.append(np.min(xx[0]))
        
#         print(xx)
#     print(al_max, al_min)
        

center_top= min(al_min)
center_bottom= max(al_max)

print(center_top, center_bottom)
print('done')
```
