```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
import os, glob
import cv2
import numpy as np
from skimage import io
from getUID import *
from get_gt import *
from roi2rect import *
from get_data_from_XML import *
from visualization_retina import *

def win_scale(data, wl, ww, dtype, out_range):
    """
    Scale pixel intensity data using specified window level, width, and intensity range.
    """
    
    data_new = np.empty(data.shape, dtype=np.double)
    data_new.fill(out_range[1]-1)
    
    data_new[data <= (wl-ww/2.0)] = out_range[0]
    
    data_new[(data>(wl-ww/2.0))&(data<=(wl+ww/2.0))] = ((data[(data>(wl-ww/2.0))&(data<=(wl+ww/2.0))]-(wl-0.5))/(ww-1.0)+0.5)*(out_range[1]-out_range[0])+out_range[0]
    
    data_new[data > (wl+ww/2.0)] = out_range[1]-1
    
    return data_new.astype(dtype)



win_dict = {'abdomen':
            {'wl': 60, 'ww': 400},
            'angio':
            {'wl': 300, 'ww': 600},
            'bone':
            {'wl': 300, 'ww': 1500},
            'brain':
            {'wl': 40, 'ww': 80},
            'chest':
            {'wl': 40, 'ww': 400},
            'lungs':
            {'wl': -400, 'ww': 1500}}
```

```python
for idx in range(len(dcm_path_list)):
    
    idx += 180
    
    
    dcm_file_list = []
    
    dicom_path = './Data/' + dcm_path_list[idx]
    dict_dcm = getUID_path(dicom_path)

    for key, dcm_path in dict_dcm.items():
#         print(dcm_path[0])
        dcm_file_list.append(dcm_path[0])

#     print(len(dict_dcm.items()))

    for i in range(len(dcm_file_list)):
        
        dcm_dcm = sitk.ReadImage(dcm_file_list[i])
        dcm_img = sitk.GetArrayFromImage(dcm_dcm)[0]

#         print(dcm_img.shape)

        jpg_path = dcm_file_list[i][:6] + '_jpg' + dcm_file_list[i][6:-4] + '.jpg'

        jpg_folder = ''

        for i in range(len(jpg_path.split('/')[:-1])):
            jpg_folder += jpg_path.split('/')[:-1][i] + '/'

#         print(jpg_folder)
#         print(jpg_path)
    
        if not os.path.exists(jpg_folder):
            os.makedirs(jpg_folder)
        
        if dcm_img.min() >= 0 and dcm_img.max() < 256:
            io.imsave(jpg_path, dcm_img)
            print(jpg_path, '  8bit image')
        else:
            
            WL = dcm_dcm.GetMetaData('0028|1050')
            WW = dcm_dcm.GetMetaData('0028|1051')

            WL = int(WL.split('\\')[0])
            WW = int(WW.split('\\')[0])
            
#             print('Window Level set : ', WL)
#             print('Window Width set : ', WW)
            
            dcm_scaled = win_scale(dcm_img, WL, WW, np.uint8, (0, 255))
    
            io.imsave(jpg_path, dcm_scaled)
            print(jpg_path, WW, WL)
```
